```python
import pycountry

# Tạo một dictionary để "nắn" các tên chuẩn ISO phức tạp về tên phổ thông gọn gàng
country_mapping = {
    "Korea, Republic of": "South Korea",
    "Viet Nam": "Vietnam",
    "Iran, Islamic Republic of": "Iran",
    "Türkiye": "Turkey",  # Hoặc giữ nguyên 'Türkiye' tùy theo phiên bản Power BI của bạn
    "Russian Federation": "Russia",
    "Taiwan, Province of China": "Taiwan"
}

def get_country_name_clean(code):
    try:
        # Lấy tên chính thức từ pycountry
        name = pycountry.countries.get(alpha_2=code).name
        # Nếu tên nằm trong danh sách cần map lại thì đổi, không thì giữ nguyên
        return country_mapping.get(name, name)
    except Exception:
        return code

# Áp dụng hàm sạch vào dataframe của bạn
df2['country_name'] = df2['country'].apply(get_country_name_clean)

# Kiểm tra kết quả
print(df2['country_name'].unique())
```

```python
df3 = df2.copy()

df3['id+'] = df3['country'].map(str) + df3['id'].map(str)
```

```python
df3.head()
```

```python
NOW = df3['date'].max() + timedelta(days=1)
NOW
```

```python
df3['days_since_purchase'] = df3['date'].apply(lambda x:(NOW - x).days)
```

```python
aggr = {
    'days_since_purchase': lambda x:x.min(),
    'date': lambda x: len([d for d in x if d >= NOW - timedelta(days=period)])
}
```

```python
rfm = df3.groupby(['id', 'id+', 'country']).agg(aggr).reset_index()
rfm.rename(columns={'days_since_purchase': 'recency',
                   'date': 'frequency'},
          inplace=True)
```

```python
rfm.drop(['id+'], axis=1, inplace=True)
```

```python
quintiles = rfm[['recency', 'frequency', 'monetary']].quantile([.2, .4, .6, .8]).to_dict()
quintiles
```

```python
def r_score(x):
    if x <= quintiles['recency'][.2]:
        return 5
    elif x <= quintiles['recency'][.4]:
        return 4
    elif x <= quintiles['recency'][.6]:
        return 3
    elif x <= quintiles['recency'][.8]:
        return 2
    else:
        return 1
```

```python
def fm_score(x, c):
    if x <= quintiles[c][.2]:
        return 1
    elif x <= quintiles[c][.4]:
        return 2
    elif x <= quintiles[c][.6]:
        return 3
    elif x <= quintiles[c][.8]:
        return 4
    else:
        return 5
```

```python
rfm['r'] = rfm['recency'].apply(lambda x: r_score(x))
rfm['f'] = rfm['frequency'].apply(lambda x: fm_score(x, 'frequency'))
rfm['m'] = rfm['monetary'].apply(lambda x: fm_score(x, 'monetary'))
```

```python
rfm.head()
```

```python
rfm['rfm_score'] = rfm['r'].map(str) + rfm['f'].map(str) + rfm['m'].map(str)
rfm.head()
```

```python
# fm = (f+m)/2
```

```python
def truncate(x):
    return math.trunc(x)
```

```python
rfm['fm'] = ((rfm['f'] + rfm['m'])/2).apply(lambda x: truncate(x))
```

Hibernating: Đã lâu không mua và mua ít.
Lost: Khách hàng đã mất, rất khó quay lại.

```python
segment_map = {
    r'22': 'hibernating',
    r'[1-2][1-2]': 'lost',
    r'15': 'can\'t lose',
    r'[1-2][3-5]': 'at risk',
    r'3[1-2]': 'about to sleep',
    r'33': 'need attention',
    r'55': 'champions',
    r'[3-5][4-5]': 'loyal customers',
    r'41': 'promising',
    r'51': 'new customers',
    r'[4-5][2-3]': 'potential loyalists'
}

rfm['segment'] = rfm['r'].map(str) + rfm['fm'].map(str)
rfm['segment'] = rfm['segment'].replace(segment_map, regex=True)
rfm.head()
```
